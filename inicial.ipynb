```python
import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
tabela.info()
display(tabela)
#object -> texto
#para sua IA aprender, ela precisa receber os dados em formato numérico! E não em texto.
```

```python
#coluna de profissão
#coluna de mix de crédito
#coluna de comportamento_pagamento
# essas colunas são textos, precisamos adaptar para elas se tornarem colunas numéricas
# a coluna que ele quer prever, pode ser em texto mesmo. (score c´redito)
#texto -> codificar para número = LabelEncoder (skit-learn)
#trocar as profissões por 1,2,3,4,5.

from sklearn.preprocessing import LabelEncoder #importando o sklearn pegando apenas o que precisa para não importar o sklearn inteiro, no caso o LabelEncoder

codificador = LabelEncoder() #cria o codificador e faaz automaticamente a adaptção de todas as profissões para números

#codificador -> aplica na coluna "profissao"
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
#codificador -> aplica na coluna "credito"
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])
#codificador -> aplica na coluna "comportamento_pagamento"
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

print(tabela.info())
```

```python
#aprendizado de máquina
#dados de treino e dados de teste
#y prevê e x usa para fazer previsão e faz a divisão de x e y
#tirar a coluna do id cliente porque é desnecessário e o score credito porque é o que vamos prever.

y = tabela["score_credito"]
x = tabela.drop(columns=["score_credito", "id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y) #esse comando de train_test_split pega sua base de dados e separa automaticamente para base de treino e base de teste.
```

```python
#criar a inteligencia artificial
#arvore da decisão -> randomforest
#Knn->  vizinhos próximos  -> vizinhos
#modelos aprendem de forma diferente, precisa saber qual modelo será bom para tal ocasião

#importar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#cria a IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#treinar o modelo
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
#testar os modelos e comparar os 
#acurácia
from sklearn.metrics import accuracy_score


previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```
